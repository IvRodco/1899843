```python
##Repaso rapido pandas

#el primer paso siempre es llamar a las librerias a utilizar
#el segundo paso es llamar los datos a usar

import pandas as pd
import numpy as np

#utilizaremos los mismos datos que la sesion pasada

cal=pd. read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/Calificaciones.csv')
cal

```

```python
#vamos a ver cuantas filas y columnas tienen nuestros datos
cal.info()
```

```python
cal=cal.rename(columns={'Calficación Final':'Cal Final'})
cal.head(5)
```

```python
# usamos iloc, seleccionamos todas las filas y accedemos a las columnas diciendole que se quede con todas menos con id
cal=cal.loc[:,cal.columns!='Id']
cal.head(5)
```

```python
cal.head()
```

```python
#el promedio del grupo

cal['Cal Final'].mean()
```

```python
#contamos todo

cal.count()
```

```python
#contamos a los alumnos

cal['Nombre'].count()
```

```python
#aprobados y reprobados

cal['situacion']=np.where(cal['Cal Final']>=70,'Aprobado','Reprobado')
```

```python
#promedio de aprobados y promedio de reprobados
cal.groupby('situacion').mean().round(1)
```

```python
#funcion agg para contar los aprobados y reprobados
cal.groupby('situacion')['Cal Final'].agg('count').round(1)
```

```python
#funcion agg para contar aprobados y reprobados, y aparte la media y varianza de cada una

cal.groupby('situacion')['Cal Final'].agg(['count','mean','std']).round()
```

# Matplotlib
Es el graficador estandar de python, es decir, muchas de las graficas de python estan hechas con esta libreria.

Existen otras librerias, pero muchas de las otras librerias estan hechas en base a matplotlib

```python
#importamos matplotlib

import matplotlib.pyplot as plt
%matplotlib inline

#vamos a hacer una grafica que nos muestre las personas y su edad

#primero vamos a hacer dos listas, una de X (personas) y una de Y(edades)
personas=[1,2,3,4]
edades=[28,21,30,27]

#le ponemos primero las etiquetas de los ejes
plt.xlabel('personas')
plt.ylabel('edades')

#le ponemos titulo al grafico
plt.title('edades de cuatro personas desconocidas')

plt.plot(personas,edades); #esta es una forma "incorrecta", para que no nos muestre el codigo de la parte superior debemos uzar .show
plt.show()
```

```python
#para un grafico de puntos


#le ponemos primero las etiquetas de los ejes
plt.xlabel('personas')
plt.ylabel('edades')

#le ponemos titulo al grafico
plt.title('edades de cuatro personas desconocidas')

plt.scatter(personas,edades)
```

```python
#Para la grafica de barras

#le ponemos primero las etiquetas de los ejes
plt.xlabel('personas')
plt.ylabel('edades')

#le ponemos titulo al grafico
plt.title('edades de cuatro personas desconocidas')

plt.barh(personas, edades)
```
