```python
from lifelines import CoxPHFitter
import seaborn as sns
```

```python
df_empleados= pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/desercion_laboral.csv')
df_empleados
```

```python
#vamos a verificar que no existan missing values

#revisamos cuantos missing values tiene cada columna
df_empleados.isnull().sum()
```

```python
#No nos sirve la variable departamento entonces la quitamos
df_empleados=df_empleados.drop('departamento',axis=1)
```

```python
#imputamos valores/los eliminamos de la base

#si los eliminamos revisamos cuantos datos perderemos
df_empleados.count()
```

```python
#volvemos a revisar que ya no haya na

df_empleados.isnull().sum()
```

```python
#Hacemos los dummies
#Modelo de Cox (en libreria lifelines) solo admite variables continuas y variables dummies(0,1)

#En la libreria lifelines las variables categoricas se tienen que convertir en dummies (con pandas)

dummies=pd.get_dummies(df_empleados['salario'], prefix='salario').astype(int)
```

```python
dummies.describe()
```

```python
#solo nos quedamos con 2 dummies
dummies=dummies[['salario_alto','salario_bajo']]
```

```python
#unimos dummies con df_empleados

df_empleados=pd.concat([df_empleados,dummies],axis=1)
df_empleados=df_empleados.drop('salario',axis=1)
```

```python
cph1=CoxPHFitter().fit(df_empleados,'satisfaccion','renunciar')
cph1.print_summary()
```

```python
#vamos a comprobar el supuesto
#H0:se viola el supuesto
#H1:se cumple el supuesto
cph1.check_assumptions(df_empleados,p_value_threshold=0.05)
```

Dado que los supuestos no cumplen con el riesgo constante, entonces como no se cumple el supuesto debemos utilizar un modelo alternativo, el cual es el Modelo de riesgo acelerado (AFT).

```python
#revisamos la distribucion del riesgo

#importamos la libreria necesaria para esto
from lifelines import ExponentialFitter, WeibullFitter, LogNormalFitter

mexpo= ExponentialFitter().fit(df_empleados['satisfaccion'],df_empleados['evaluacion'])
mweibul=WeibullFitter().fit(df_empleados['satisfaccion'],df_empleados['evaluacion'])
mlogn=LogNormalFitter().fit(df_empleados['satisfaccion'],df_empleados['evaluacion'])

mexpo.AIC_ , mweibul.AIC_ , mlogn.AIC_
```

```python
#Dado que con el AIC se elige el de menor valor, eligiremos el Modelo Weibull

#importamos la libreria para el AFT
from lifelines import WeibullAFTFitter

weibulAFT=WeibullAFTFitter().fit(df_empleados,'satisfaccion','renunciar')
weibulAFT.print_summary()
```
