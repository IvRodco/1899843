```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#libreria lifelines

from lifelines import ExponentialFitter
```

```python
#traemos los datos, que son datos sobre recaidas
Datos_rec= pd.read_csv('https://raw.githubusercontent.com/jimmyzac/Estadistica-Aplicada-FCFM-UANL/main/bases_datos/ejercicio_expo.csv',index_col=0)
Datos_rec
```

```python
##exp= ExponentialFitter().fit(tiempo,censura)
exp= ExponentialFitter().fit(Datos_rec[' tiempo'], Datos_rec[' estado'])

##grafico de la funcion de supervivencia
exp.plot_survival_function();
plt.title('funcion de supervivencia Exponencial($\lambda$)');
```

```python
#Riesgo
exp.hazard_
```

```python
#El valor de la supervivencia en un tiempo especifico
exp.survival_function_at_times(16)
```

```python
#grafique la funcion de riesgo (tiene confianza del 95%)
exp.plot_hazard();
plt.title('funcion de riesgo Exponencial  $\lambda=0.033$');
```
