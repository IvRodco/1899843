```python
#importamos librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

#libreria MCO
import statsmodels.formula.api as smf
```

```python
datos= pd.read_excel('modelolineal.xlsx')
datos
```

```python
##aparentemente existe una relacion lineal entonces se puede estimar minimos cuadrados ordinarios
#se puede estimar MCO

reg= smf.ols(formula='Y~X', data='datos')
results= reg.fit()
print(results.rummary())
```

```python
#visto tenemos una prueba de hipotesis en donde H0:B1=0 y H1:B1=/=0
##Intercepto: p value<alfa -----> se rechaza H0, significa que el intercepto es distintode cero
##(Xi|Bi)---> p value<alfa------> rechazamos H0, significa que X si afecta Y
##En promedio cuando X incrementa en uno Y incrementa en 0.58
```
