Para evaluar un modelo generalmente se usa $R^2$ .

Pero no es la unica medida utilizada, tambien se suele usar con bastante precuencia el MAD, MAPE y MSD.



MAD: Desviacion Absoluta Media

MAPE:Error Porcentual Absoluto Medio

MSD: Desviacion Cuadratica Media


$MAD=1/n∑|(Y-\hat(Y))|$

$MAPE=1/n∑|((Y-\hat(Y))/Y)|$

$MAS=1/n∑(Y-\hat(Y))^2$
