##Pandas

Se utiliza para manipulas datos estructurados.

Pandas esta construida en base a numpy, eso significa que cualquier funcion de nunpy funciona a en pandas.

En Pandas hay 3 estructuras de datos:
1. Series (unidimencionales)
2. Dataframes (datos en filas y columnas)
3. Panel (se organizan con el tiempo)




Supongamos que conocemos las edades de la siguientes personas:
Juan 10 años, Pedro 20 años, Jorge 30 años, Ale 40 años y Alba 50 años.

```python
Edades=[10,20,30,40,50]

Edades
```

##Dataframe

los dataframe son datos con filas y columnas

```python
alumno1=['Juan','Perez',23]
alumno2=['Raul','Gonzalez',17]
alumno3=['Jimena','Rosas',20]
alumno4=['Alejandra','Saldaña',21]

#hacemos una lista que contenga a los alumnos
clase_estadistica=[alumno1,alumno2,alumno3,alumno4]
clase_estadistica
```

```python
#creamos el dataframe
df_clase= pd.DataFrame(clase_estadistica,columns=['Nombre','Apellido','Edad'])
type(df_clase)
```

```python
df_clase
```

```python
#nos arroja como estasn las filas
df_clase.index
```

```python
#nos arroja el nombre de las columnas
df_clase.columns
```

```python
df_clase['Estado']=['Nuevo Leon','Coahuila','Nuevo Leon','Verecruz']
df_clase
```

```python

df_clase=df_clase.set_index('Nombre')
df_clase
```

```python
df_clase['Edad en dias']=df_clase['Edad']*365
df_clase
```

```python
df_clase['Edad en meses']=df_clase['Edad']*12
df_clase
```

##Seleccion en Dataframe

1. Seleccion estandar
2. Seleccion de iloc
3. Seleccion loc
4. Seleccion por boleanos (filtros)

```python
# Seleccion estandar

 #seleccionamos solo una columna
 #en este caso seleccionamos apellido
df_clase['Apellido']
```

```python
 #ahora seleccionamos la columna edad y apellido
df_clase[['Edad','Apellido']]
```

```python
#Ahora seelccionamos por filas

#en esta caso seleccionaremos fila de la 0 a la 2
df_clase[0:2]
```

```python
#ahora eligiremos filas y columnas
#Primero de seleccionan las filas y luego las columnas

#En este caso seleccionamos las filas de la 0 a la 2 y la columna edad
df_clase[0:2]['Edad']
```

```python
#Ahora seleccionamos las dilas de la 0 a la 2 y la columnas edas y apellido

df_clase[0:2][['Edad','Apellido']]
```

```python
#Seleccion por posicion iloc
#

df_clase.iloc[0:2,1]
```

```python
#quiero seleccionar solo la columa, sin fila

df_clase.iloc[:,1]
```

```python
#seleccion por etiquetas loc
#dataframe.loc[filas,columnas]

df_clase.loc[:,'Edad']
```

```python
df_clase.loc['Juan','Edad']
```

```python
#seleccion de condicionales(boleanos)
df_clase[df_clase['Edad']>20]
```

```python
#para el promedio de las edades
df_clase['Edad'].mean()
```

```python
#para la desv estandar de las edades
df_clase['Edad'].std()
```

```python
#para la suma de las edades
df_clase['Edad'].sum()
```

```python
#funcion si
#np.where(condicion, valor si se cumple,valor si no se cumple)

import numpy as np

df_clase['Mayor de edad']= np.where(df_clase['Edad']>=18,'Es mayor de edad','Menor de edad')
df_clase
```

```python
#funcion si conjunto
#np.select(condiciones,valor si se cumple cada condicion,valor si no se cumple una condicion)
condiciones=[df_clase['Edad']<=17,
             df_clase['Edad']<23,
             df_clase['Edad']>=23]
valor_condicion=['Prepa','Licenciatura','Egresado']
df_clase['Condicion de estudio']= np.select(condiciones,valor_condicion,default='Dato invalido')
df_clase
```

```python
#.count cuenta cuantos datos tenemos en cierta columna
df_clase['Edad'].count()
```

```python
#ahora contaremos cuantos son mayores de edad y cuentos son menores de edad

df_clase.value_counts('Mayor de edad')
```

```python
#ahora contaremos cuantos son de cada grado de estudio

df_clase.value_counts('Condicion de estudio')
```

```python
#vamos a sacar los promedio de cada agrupacion que hicimos

#primero los acomodamos por la columna mayor de edad
df_clase.groupby('Mayor de edad').mean()
```

```python
#ahora vamos a seleccionar solo una columna, que sera la de edad
df_clase.groupby('Mayor de edad').mean().round(1)['Edad']
```

```python
#agg
#sirve para resumir

#ahora vamos a resumir cuantos hay en cada columna
df_clase.agg(['count'])
```

```python
#ahora vamos a resumir cuantos hay en cada columna con la media
df_clase.agg(['count','mean'])
```

```python
#selaccionar a alejandra
df_clase.query('Nombre=="Alejandra"')
```
