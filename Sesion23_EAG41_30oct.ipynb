```python
#Basicas
import pandas as pd
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

#especializadas

from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, plot_predict
from statsmodels.tsa.arima.model import ARIMA 

#pmdarima
import pmdarima as pm


```

```python
euro=pd.read_excel('euro_mxn.xlsx')
euro.head(5)
```

```python
#tipo de variables
euro.dtypes
```

```python
#grafica de la serie
euro['tipo_euro'].plot()
plt.title('Tipo de cambio euro-peso')
plt.xlabel('semana')
plt.ylabel('pesos x euro')
plt.show();
```

```python
#prueba de raiz unitaria
prueba_adf=adfuller(euro['tipo_euro'])
pvalue=prueba_adf[1]
pvalue.round(2)
```

H0:Serie no estacionaria
H1:Serie estacionaria

=> No rechazamos H0, entonces el euro es una serie no estacionaria

```python
#primeras diferencias
euro['euro_diff']=euro['tipo_euro'].diff()
```

```python
#Graficar primeras diferencias
euro['euro_diff'].plot()
plt.title('Tipo de cambio euro-peso')
plt.xlabel('semana')
plt.ylabel('pesos x euro')
plt.show();
```

```python
#prueba adf primeras diferencias

prueba_adf_diff=adfuller(euro['euro_diff'].dropna())
pvalue_diff=prueba_adf_diff[1]
pvalue_diff.round(2)
```

H0: no estacionaria
H1: estacionaria
Rechazamos H0, entonces las primeras diferencias son un proceso estacionario

```python
#graficas de correlacion

ig,ax=plt.subplots(1,2,figsize=(10,4))

#1
plot_acf(euro['euro_diff'].dropna(),zero=False,lags=30,ax=ax[0])

#2
plot_pacf(euro['euro_diff'].dropna(),zero=False,lags=30,ax=ax[1])

plt.show();
```

Esta serie no tiene componente estacional

```python
#determinar orden del proceso

arimaorder_aic=pm.auto_arima(euro['tipo_euro'],
                             start_p=1,start_q=1,
                             max_p=7,max_q=7,
                             information_criterion='aic')
arimaorder_aic.params
```

Como es un ruido blanco, no se puede pronosticar ya que no tiene memoria

```python
#hacemos if
# vamos a estimar varios modelos y que se guarde el AIC BIC
ic = {}
for ar in range(7):
    for ma in range(7):
        print(f"AR: {ar}, MA: {ma}")
        modelo = sm.tsa.arima.ARIMA(euro['tipo_euro'].dropna(), order=(ar, 1, ma))
        resultado = modelo.fit()
        ic[(ar, ma)] = [resultado.aic, resultado.bic]

ic = pd.DataFrame(ic, index=["AIC", "BIC"]).T
ic.index = ic.index.set_names(["AR", "MA"])
```

```python
#revisamos que tipo de arima sugiere
ic.sort_values('AIC').head()
```

Como no se puede 0 y 0 es ruido blanco y no se puede pronosticar, entonces el euro es una arima(1,1,0)

```python
#
ic.sort_values('AIC').head()
#sugiere lo mismo que el anterior
```

```python
#estimamos modelo
model=ARIMA(euro['tipo_euro'],order=(1,1,0))
results=model.fit()
print(results.summary())
```

--------------ARIMA(1,1,0)----------------
el primer uno es la p, luego el orden de integracion, luego q

El modelo es el que dice ar.L1

Entonces con los resultados son:
 p=0.036
 q=0.1536

```python
#hacemos estimaciones escritas con el intervalo de confianza
forecast=results.get_forecast(steps=4)
forecast_mean=forecast.predicted_mean
forecast_ci=forecast.conf_int()
```

```python
#vamos hasta que semana llega
euro.tail()
```

```python
#establecemos cuantas semanas puede predecir
inicio='2023-10-30'
fin='2023-11-20'
```
