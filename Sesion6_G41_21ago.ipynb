```python
#importamos librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
nacimientos= pd.DataFrame({'municipio':['Mty','Apodaca','Escobedo','Santiago'],'bb_nacidos':[233,569,780,123],'Defunciones':[123,80,23,12]})
nacimientos
```

```python
nacimientos= nacimientos.sort_values('bb_nacidos',ascending=False)
```

```python
plt.bar(nacimientos['municipio'],nacimientos['bb_nacidos'])
plt.title('Nacimientos de N.L 2023')
plt.xlabel('Municipios')
plt.ylabel('Cantidad de bebes nacidos')
plt.show()
```

```python

personas=['Juan','Pedro','Ana','Ale']
edades=[23,18,20,31]

plt.bar(personas,edades)
plt.show()
```

```python
#Vamos a ver otra forma de graficar, con un tipo lienzo

#fig es el nombre del lienzo
#aqio decimos que se acomode en dos filas y una columna
fig,(panel1,panel2)=plt.subplots(2,1)

#aqui se pode que ambas columnas sean de igual dimensiones(igual de ancho y alto)
fig.tight_layout()

#comenzamos con la grafica 1
x1= nacimientos['municipio']
y1= nacimientos['bb_nacidos']
panel1.bar(x1,y1)

#Ahora grafica 2
x2= personas
y2= edades
panel2.bar(x2,y2)

plt.show()
```
